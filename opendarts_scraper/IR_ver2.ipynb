```python
def extract_zip_file(zip_path, extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
```

```python
def extract_files_with_keyword(directory, keyword, extension):
    matching_files = []
    for file in os.listdir(directory):
        if keyword in file and file.endswith(extension):
            matching_files.append(file)
    return matching_files
```

```python
def get_exception_list(file_list, pattern):
    exception_list = []
    for file in file_list:
        match = re.search(pattern, file)
        if match:
            first_bracket = match.group(1)
            exception_list.append(file)
    return exception_list
```

```python
def get_revised_values(html_content):
    values = []
    pattern_recog = r'<span style="width:242px;font-size:10pt;">1\. 배당구분<\/span>\s*<\/td>\s*<td colspan="2" width="352"> <span class="xforms_input" style="width:352px;font-size:10pt;">(.*?)<\/span>'
    pattern_cate = r'<span style="width:242px;font-size:10pt;">2. 배당종류<\/span>\s*<\/td>\s*<td colspan="2" width="352"> <span class="xforms_input" style="width:352px;font-size:10pt;">(.*?)<\/span>'

    match_recog = re.search(pattern_recog, html_content)
    match_cate = re.search(pattern_cate, html_content)
    
    if match_recog:
            value = match_recog.group(1)
            values.append(value)
    else:
        values.append("")
            
            
    if match_cate:
        value_two = match_cate.group(1)
        values.append(value_two)
    else:
        values.append("")

    return values


```

```python
def click_next_page(start, end):
    for i in range(start, end+1):
        if i > start:
            # <a> 태그 찾기
            a_element = wait.until(EC.presence_of_element_located((By.XPATH, f'//a[text()="{i}"]')))

            # <a> 태그 클릭
            ActionChains(driver).move_to_element(a_element).click().perform()

            # 다음 페이지 로드 대기
            wait.until(EC.staleness_of(a_element))

        # 추가로 수행할 코드
        tbody = driver.find_element('id', 'tbody')
        td_elements = tbody.find_elements('css selector', 'td.tL')
        id_values = []

        for td_element in td_elements:
            a_tags = td_element.find_elements('tag name', 'a')
            if a_tags and a_tags != '':
                a_tag = a_tags[0]
                id_values.append(a_tag.get_attribute('id'))

        for id in range(1, len(id_values), 2):
            find_elementss('id', id_values[id])

        # 다운로드 파일 처리
        for j in range(1, len(driver.window_handles)):
            # 스위치
            driver.switch_to.window(driver.window_handles[j])
            print(driver.current_window_handle)

            # 다운로드 버튼 클릭
            download_button = driver.find_element(By.CLASS_NAME, 'btnDown')
            download_button.click()

            # 새 창으로 전환
            driver.switch_to.window(driver.window_handles[-1])

            # 링크 클릭
            link = driver.find_element(By.CSS_SELECTOR, 'a.btnFile')
            href_value = link.get_attribute('href')
            link.click()

            # 대기
            time.sleep(3)

        # 핸들 변경
        main = driver.window_handles
        for k in main:
            if k != main[0]:
                driver.switch_to.window(k)
                driver.close()
        driver.switch_to.window(driver.window_handles[0])
```

```python
patterns = [
    r'<span class="xforms_input" style="width:\d+px;font-size:10pt;">(\d{4}-\d{2}-\d{2})</span>',
    r'<span style="width:191px;font-size:10pt;">2. 장소</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>',
    r'<span style="width:191px;font-size:10pt;">3. 대상자</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>',
    r'<span style="width:191px;font-size:10pt;">4. 실시목적</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>',
    r'<span style="width:191px;font-size:10pt;">5. 실시방법</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>',
    r'<span style="width:191px;font-size:10pt;">6. 주요내용</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>',
    r'<span style="width:78px;font-size:10pt;text-align:center;">담당부서\(담당자\)</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>',
    r'<span style="width:78px;font-size:10pt;text-align:center;">전화번호\(팩스번호\)</span> </td> \n       <td colspan="4" width="406"> <span class="xforms_input" style="width:406px;font-size:10pt;">(.*?)</span>'
]
```

```python
# result_list 값들을 열 이름에 맞게 구성
column_names = ['0. 회사명', '1. 일시', '2. 장소', '3. 대상자', '4. 실시 목적', '5. 실시 방법', '6. 주요내용', '7. 담당 부서', '8. 전화번호']

# 데이터프레임 생성
df_output = pd.DataFrame(columns=column_names)

for filename in file_list:
    file_path = os.path.join(zip_path, filename)
    with codecs.open(file_path, 'r') as test_f:
        text_test = test_f.read()
        
        # 변수 초기화
        name = ''
        date = ''
        location = ''
        target = ''
        purpose = ''
        method = ''
        content = ''
        department = ''
        phone_number = ''

        # 0. 회사명
        patter_cor = r'<title>(.*?)</title>' 
        matches_cor = re.findall(patter_cor, text_test)
        pattern_cor_revised = r'^(.*?)\/'
        result_name = re.findall(pattern_cor_revised, matches_cor[0])
        if result_name:
            name = result_name[0]
           
        else:
            print("일치하는 패턴을 찾을 수 없습니다.1")
    
        for pattern in patterns:
            # get values 
            if pattern == patterns[0]:
                match = re.search(pattern, text_test)
                if match:
                    date = match.group(1)

            elif pattern == patterns[1]:
                match = re.search(pattern, text_test)
                if match:
                    location = match.group(1)

            elif pattern == patterns[2]:
                match = re.search(pattern, text_test)
                if match:
                    target = match.group(1)

            elif pattern == patterns[3]:
                match = re.search(pattern, text_test)
                if match:
                    purpose = match.group(1)

            elif pattern == patterns[4]:
                match = re.search(pattern, text_test)
                if match:
                    method = match.group(1)

            elif pattern == patterns[5]:
                match = re.search(pattern, text_test)
                if match:
                    result = match.group(1)
                    clean_result = re.sub('<.*?>', '', result)
                    clean_result = clean_result.strip()
                    korean_only = re.findall('[가-힣]+', clean_result)
                    content = ' '.join(korean_only)

            elif pattern == patterns[6]:
                match = re.search(pattern, text_test)
                if match:
                    department = match.group(1)

            elif pattern == patterns[7]:
                match = re.search(pattern, text_test)
                if match:
                    phone_number = match.group(1)
                    
        # 데이터프레임에 행 추가
        df_output = pd.concat([df_output, pd.DataFrame([[name, date, location, target, purpose, method, content, department, phone_number]], columns=column_names)], ignore_index=True)

# 데이터프레임 출력
print(df_output)
```
